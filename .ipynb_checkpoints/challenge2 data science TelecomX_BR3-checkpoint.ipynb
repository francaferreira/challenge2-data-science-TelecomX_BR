Relatório de Análise de Evasão de Clientes da TelecomX
1. Introdução
A evasão de clientes (Churn) é um dos principais desafios enfrentados por empresas de telecomunicações. Quando um cliente cancela seu serviço, a empresa não só perde receita, mas também enfrenta custos adicionais para adquirir novos clientes.

Objetivo:

Realizar ETL (Extração, Transformação e Limpeza) dos dados da TelecomX.

Identificar padrões e possíveis causas da evasão de clientes.

Gerar visualizações para auxiliar na tomada de decisão.

Preparar os dados para análises preditivas futuras que possam reduzir a taxa de churn.

2. Processo de ETL (Extração, Transformação e Limpeza)
2.1. Extração dos Dados
Os dados foram obtidos de um arquivo JSON hospedado no GitHub, contendo informações sobre clientes da TelecomX, como:

Dados demográficos (gênero, idade, dependentes)

Serviços contratados (telefonia, internet, TV)

Informações financeiras (tipo de contrato, forma de pagamento, cobrança mensal)

Status de Churn (se o cliente cancelou ou não).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json'
df = pd.read_json(url)
```

```python
df
```

2.2. Normalização dos Dados Aninhados
Os dados estavam estruturados em colunas aninhadas (JSON), então foi necessário desnormalizá-los para análise:

```python
# Normalizar colunas JSON
df_customer = pd.json_normalize(df['customer'])
df_phone = pd.json_normalize(df['phone'])
df_internet = pd.json_normalize(df['internet'])
df_account = pd.json_normalize(df['account'])

# Concatenar em um único DataFrame
df_final = pd.concat([
    df[['customerID', 'Churn']],
    df_customer,
    df_phone,
    df_internet,
    df_account
], axis=1)
```

```python
df_final
```

2.3. Limpeza e Tratamento
Verificação de valores nulos:

```python
print(df_final.isnull().sum())
```

Resultado: Nenhum dado faltante.

# Conversão de variáveis categóricas:

```python
# Codificar 'Churn' (Yes/No → 1/0)
df_final['Churn'] = df_final['Churn'].map({'Yes': 1, 'No': 0})
```

```python
df_final
```

```python
df_final.info()
```

```python
# Passo 1: Tentar conversão para float
df_final['Charges.Total'] = pd.to_numeric(df_final['Charges.Total'], errors='coerce')
```

```python
print(df_final['Charges.Total'].unique())
```

```python
print(df_final.describe())
```

## 3. Análise Exploratória (EDA)
### 3.1. Distribuição de Churn

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Churn', data=df_final, palette='viridis')
plt.title('Distribuição de Churn (Evasão de Clientes)')
plt.xlabel('Churn (0 = Não, 1 = Sim)')
plt.ylabel('Contagem de Clientes')
plt.show()
```

Insight:

25,7% dos clientes cancelaram (1.869 casos).

74,3% permaneceram (5.174 casos).



```python
# Variáveis numéricas
numeric_cols = ['SeniorCitizen', 'tenure', 'Charges.Monthly', 'Charges.Total']
```

```python
# Boxplot para variáveis numéricas vs churn
plt.figure(figsize=(15, 5))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(1, 4, i)
    sns.boxplot(x='Churn', y=col, data=df_final)
    plt.title(f'Distribuição de {col} por Churn')
plt.tight_layout()
plt.show()
```

## 3.2. Principais Fatores Relacionados ao Churn
### a) Tipo de Contrato vs. Churn

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Contract', hue='Churn', data=df_final, palette='viridis')
plt.title('Churn por Tipo de Contrato')
plt.xlabel('Contagem de Clientes')
plt.ylabel('Tipo de Contrato')
plt.legend(title='Churn', labels=['Não', 'Sim'])
plt.show()
```

Insight:

Clientes com contratos mensais têm maior taxa de evasão.

Contratos anuais/bienais retêm mais clientes.

### b) Tempo de Cliente (Tenure) vs. Churn

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Churn', y='tenure', data=df_final, palette='viridis')
plt.title('Tempo de Cliente (Tenure) vs. Churn')
plt.xlabel('Churn (0 = Não, 1 = Sim)')
plt.ylabel('Meses como Cliente')
plt.show()
```

Insight:

Clientes que cancelam tendem a ter menos tempo de serviço.

Clientes antigos são mais fiéis.

### c) Forma de Pagamento vs. Churn

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='PaymentMethod', hue='Churn', data=df_final, palette='viridis')
plt.title('Churn por Método de Pagamento')
plt.xlabel('Método de Pagamento')
plt.ylabel('Contagem de Clientes')
plt.legend(title='Churn', labels=['Não', 'Sim'])
plt.xticks(rotation=45)
plt.show()
```

Insight:

Clientes que pagam por cheque eletrônico têm maior evasão.

Pagamentos automáticos (cartão/transferência) reduzem churn.

# 4. Conclusão e Recomendações

## Principais Achados

Contratos mensais têm maior evasão → Incentivar planos anuais.

Clientes novos cancelam mais → Melhorar onboarding e fidelização.

Pagamentos manuais (cheque eletrônico) aumentam churn → Promover débito automático.
